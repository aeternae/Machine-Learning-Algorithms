```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import tensorflow as tf
```

## 回归问题

```python
boston = datasets.load_boston()
```

```python
X = boston.data
y = boston.target
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
standardScaler.fit(X)
X = standardScaler.transform(X)
```

```python
df = pd.DataFrame(X, y, columns=boston.feature_names)
df.head()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=111)
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(20, input_shape=(13, ), activation='relu'))
# model.add(tf.keras.layers.Dense(10, input_shape=(20, ), activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='mse')
```

```python
model.fit(X_train, y_train, epochs=200)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pre = model.predict(X_test)
```

## 二分类问题

```python
cancer = datasets.load_breast_cancer()
```

```python
X = cancer.data
y = cancer.target
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
standardScaler.fit(X)
X = standardScaler.transform(X)
```

```python
df = pd.DataFrame(X, y, columns=cancer.feature_names)
df.head()
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=111)
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(20, input_shape=(30,), activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics=['acc'])
```

```python
history = model.fit(X_train, y_train, epochs=200)
```

```python
history.history.keys()
```

```python
plt.plot(history.epoch, history.history.get('loss'))
```

```python
plt.plot(history.epoch, history.history.get('acc'))
```

```python
model.evaluate(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=500, random_state=666, oob_score=True, n_jobs=-1)
rf_clf.fit(X_train, y_train)
```

```python
rf_clf.score(X_test, y_test)
```
